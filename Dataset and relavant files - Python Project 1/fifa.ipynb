```python
df_fifa['Value']=np.where(df_fifa['Value'].str[-1:]=='M',pd.to_numeric(df_fifa['Value'].str[:-1]) * 1000000, 
        np.where(df_fifa['Value'].str[-1:]=='K',pd.to_numeric(df_fifa['Value'].str[:-1]) * 1000, 
        pd.to_numeric(df_fifa['Value'].str[:-1])))
```

```python
df_fifa['Wage']=np.where(df_fifa['Wage'].str[-1:]=='M',pd.to_numeric(df_fifa['Wage'].str[:-1]) * 1000000, 
        np.where(df_fifa['Wage'].str[-1:]=='K',pd.to_numeric(df_fifa['Wage'].str[:-1]) * 1000, 
        pd.to_numeric(df_fifa['Wage'].str[:-1])))
```

```python
df_fifa['Release Clause']=np.where(df_fifa['Release Clause'].str[-1:]=='M',pd.to_numeric(df_fifa['Release Clause'].str[:-1]) * 1000000, 
        np.where(df_fifa['Release Clause'].str[-1:]=='K',pd.to_numeric(df_fifa['Release Clause'].str[:-1]) * 1000, 
        pd.to_numeric(df_fifa['Release Clause'].str[:-1])))
```

20. Generate a data frame containing top 5 players by Overall rating for
each unique position. (2 Points)


```python
Top5ofEachPosition=df_fifa.groupby(['Position','Name'])['Overall'].mean().reset_index().sort_values(by = ['Position','Overall'],ascending=False)
```

```python
Top5ofEachPosition['rank'] = 1
```

```python
Top5ofEachPosition['Rank2']=Top5ofEachPosition.groupby('Position')['rank'].cumsum()
```

```python
Top5ofEachPosition.drop(['rank'],axis=1, inplace=True)
```

```python
Top5ofEachPosition=Top5ofEachPosition[Top5ofEachPosition['Rank2']<=5]
```

21. What is the average wage one can expect to pay for the top 5 players
in every position? (use the data frame created in Q19) (2 points)

```python
Top5ofEachPosition=df_fifa.groupby(['Position','Name'])['Overall','Wage'].mean().reset_index().sort_values(by = ['Position','Overall'],ascending=False)
Top5ofEachPosition['rank'] = 1
Top5ofEachPosition['Rank2']=Top5ofEachPosition.groupby('Position')['rank'].cumsum()
Top5ofEachPosition.drop(['rank'],axis=1, inplace=True)
Top5ofEachPosition=Top5ofEachPosition[Top5ofEachPosition['Rank2']<=5]
```

```python
chart = plt.figure(figsize=(16,8))

chart=sns.barplot(x = 'Position',
            y =  'Wage',
            data = Top5ofEachPosition)

chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=90, 
    horizontalalignment='right'
)

plt.show()
```
